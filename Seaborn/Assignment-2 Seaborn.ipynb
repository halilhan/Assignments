## 2: Plot 12 most dangerous cities

```python
dangerous_cities = kill.city.value_counts().head(12)
dangerous_cities
```

```python
dangerous_cities_index = dangerous_cities.index
dangerous_cities_index
```

```python
# with seaborn barplot
plt.figure(figsize=(8,5))
sns.barplot(x=dangerous_cities_index, y= dangerous_cities)
plt.xticks(rotation= 45)
plt.title('Most Dangerous Cities', color='b', fontsize= 16)
plt.show()
```

```python
# with seaborn countplot
plt.figure(figsize=(8,5))
sns.countplot(x='city', order= dangerous_cities_index, data=kill)
plt.xticks(rotation= 45)
plt.title('Most Dangerous Cities', color='b', fontsize= 16)
plt.show()
```

## 3: Plot 20 most dangerous states

```python
dangerous_state = kill.state.value_counts().head(20)
dangerous_state
```

```python
dangerous_state_index = dangerous_state.index
dangerous_state_index
```

```python
# with seaborn barplot
plt.figure(figsize=(8,5))
sns.barplot(x=dangerous_state_index, y=dangerous_state)
plt.xticks(rotation = 45)
plt.title('Most Dangerous States', fontsize=16, color='b')
plt.show()
```

```python
# with seaborn countplot
plt.figure(figsize=(8,5))
sns.barplot( x=dangerous_state_index, y=dangerous_state)
plt.xticks(rotation = 45)
plt.title('Most Dangerous States', fontsize=16, color='b')
plt.show()
```
